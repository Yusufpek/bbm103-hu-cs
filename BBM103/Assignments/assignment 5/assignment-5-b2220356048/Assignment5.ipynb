<h5>Convert to values in csv file to numpy array for calculations</h5>

```python
allValues = df.values
allValues
```

<h5>Convert to personality types to integers</h5>

```python
PERSONALITIES = {
"ESTJ": 0,
"ENTJ": 1,
"ESFJ": 2,
"ENFJ": 3,
"ISTJ": 4,
"ISFJ": 5,
"INTJ": 6,
"INFJ": 7,
"ESTP": 8,
"ESFP": 9,
"ENTP": 10,
"ENFP": 11,
"ISTP": 12,
"ISFP": 13,
"INTP": 14,
"INFP": 15
}

print(allValues[-1][-1])
for person in PERSONALITIES:
    allValues[allValues == person] = PERSONALITIES[person]
print(allValues[-1][-1])
```

<h4>Feature Normalization</h4> <br/> 

![Formula](https://i.ibb.co/C7BM3JW/image.png)

```python
valuesWithoutPersonality = allValues[:, :-1] # drop the personalities
normalizatedValues = (valuesWithoutPersonality - np.min(valuesWithoutPersonality)) / (np.max(valuesWithoutPersonality) - np.min(valuesWithoutPersonality))
print(normalizatedValues)
normalizatedValues = np.column_stack((normalizatedValues, allValues[:, -1:])) # add the personality to end of the array
print(normalizatedValues)
```

Get the test and train values, shuffle the list for working with different data.

```python
def getValues(isNormalize = True, kFold = 1):
    global allValues
    if(isNormalize): allValues = normalizatedValues # normalizated
    values = allValues[ (kFold - 1) * 12000 :12000 * kFold]
    beforeTest = allValues[:(kFold - 1) * 12000]
    afterTest = allValues[12000 * kFold:]
    knownValues = np.row_stack((beforeTest , afterTest))
    print(f"Test Shape: {values.shape}")
    print(f"Train Shape: {knownValues.shape}")
    return values, knownValues
```

Calculating differences, finding nearest neighbor for given test, train values for giving K-values of list

```python
def findPersonalityWithKNN(values, knownValues, K):
    personalities = []
    personalitiesK = []
    true = 0
    false = 0
    b = np.array(knownValues[:,:-1], dtype=np.float32)
    for value in values:
        a = np.array([value[:-1]], dtype=np.float32)
        distances = np.linalg.norm(a-b, axis=1)
        distances = np.column_stack((distances, knownValues[:,-1:]))
        distances = sorted(distances, key=lambda entry: entry[0])
        for kv in K:
            firstKdist = distances[:kv]
            distancesCounts = {}
            for d in firstKdist:
                try:
                    distancesCounts[d[1]] += 1
                except:
                    distancesCounts[d[1]] = 1
            distancesCounts =  dict(sorted(distancesCounts.items(), key=lambda item: item[1], reverse=True))
            personality = list(distancesCounts.keys())[0]
            personalitiesK.append(personality)
        pk = personalitiesK.copy()
        personalities.append(pk)
        personalitiesK.clear()
    updatedValues = np.column_stack((values, personalities))
    print("Add the founded nearest personalities to array...")
    print(f"Updated array shape: {updatedValues.shape}")
    return updatedValues
```

<h5>Calculation of the accuracy, precision and recall values for given data.</h5>
<img src="https://i.ibb.co/938q35N/image.png" alt="accuracy" style="height:100px;"/> <img src="https://i.ibb.co/J5B4XnR/image.png" alt="precision" style="height:100px;"/> <img src="https://i.ibb.co/hH8jJ2q/image.png" alt="recall" style="height:100px;"/>

Look for the founded personalities with knn and original personalities.</br>
True negative value is equal to sum of other personalities true positive values.

```python
def findAccuracyOfValues(updatedValues, K, isNormalize):
    performanceEvulation = {y: {i: { "TP" : 0, "FP" : 0, "FN" : 0, "TN" : 0} for i in range(16) } for y in range(len(K))}
    
    #find the true positive, flase negative and false pozitive counts
    for k in range(len(K)):
        for value in updatedValues:
            if(value[-(len(K) + 1)] == value[-(len(K) - k)]):
                performanceEvulation[k][value[-(len(K) + 1)]]["TP"] += 1
            else:
                performanceEvulation[k][value[-(len(K) + 1)]]["FN"] += 1
                performanceEvulation[k][value[-(len(K) - k)]]["FP"] += 1
    # calculate the true negative count with addition of all other values true positive counts
    for k in range(len(K)):
        for key in performanceEvulation[k]:
            for key2 in performanceEvulation[k]:
                if key == key2: continue
                performanceEvulation[k][key]["TN"] += performanceEvulation[k][key2]["TP"]
    accuracies = []
    precisions = []
    recalls = []
    macroAccuracies = []
    # find the accuracy, precision, recall values. calculate them in micro average and macro average. 
    for k in range(len(K)):
        print(f"K: {2 * (k) + 1} \t\t\t Is Normalize: {isNormalize}")
        print("Micro: ")
        for key, v in PERSONALITIES.items():
            TP =  performanceEvulation[k][v]["TP"] 
            TN =  performanceEvulation[k][v]["TN"]
            FP =  performanceEvulation[k][v]["FP"]
            FN =  performanceEvulation[k][v]["FN"]
            accuracy = (TP + TN) / (TP + TN + FP + FN)
            precision = TP / (TP + FP)
            recall = TP / (TP + FN)
            accuracies.append(accuracy)
            precisions.append(precision)
            recalls.append(recall)
            print(f"{key}-> Accuracy: % {accuracy * 100} \t Precision: % {precision * 100} \t Recall: % {recall * 100}")    

        macroAccuracies.append( "% {:.2f}".format(sum(accuracies) / len(accuracies) * 100))
        print("\nMacro: ")
        print(f"Accuracies: {macroAccuracies[-1]}")
        print("Precision: % {:.2f}".format(sum(precisions) / len(precisions) * 100))
        print("Recall: % {:.2f}\n".format(sum(recalls) / len(recalls) * 100))
        accuracies = []
        precisions = []
        recalls = []
    #print a list which has all macro accuracy values
    for i in range(len(K)):
        print(f"K={K[i]}-> Accuracy: {macroAccuracies[i]}")
    print("") # last line
```

<h4>Time and Accuracy Comparison</h4> (Different k values and normalization)

Define a function for calculation

```python
def calculate(isNormalize, K):
    global updatedValues
    values, knownValues = getValues(isNormalize=isNormalize, kFold= 1)
    updatedValues = findPersonalityWithKNN(values, knownValues, K = K)
    findAccuracyOfValues(updatedValues, K, isNormalize=isNormalize)
```
